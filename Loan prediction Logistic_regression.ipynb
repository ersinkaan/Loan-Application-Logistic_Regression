```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
loan_test = pd.read_csv('loan_test.csv')
loan_train = pd.read_csv('loan_train.csv')
```

```python
loan_test.head()
```

```python
#handling missing values on the training data
loan_train.isnull().sum()
for i in loan_train.isnull().sum():
    if i == False:
        print('There is {} missing values in this feature, handling is not required!'.format([i]))
    else:
        print('There is {} missing values, handling is required'.format([i]))
```

```python
#handling missing values on the test data
loan_test.isnull().sum()
for i in loan_test.isnull().sum():
    if i == False:
        print('There is {} missing values in this feature, handling is not required!'.format([i]))
    else:
        print('There is {} missing values, handling is required'.format([i]))
```

```python
#I will do missing values handling for practice
loan_train["Loan_Status"].fillna(loan_train["Loan_Status"].mode()[0],inplace=True)
loan_train["Married"].fillna(loan_train["Married"].median(),inplace=True)
loan_train["Education"].fillna(loan_train["Education"].median(),inplace=True)
loan_train["Self_Employed"].fillna(loan_train["Self_Employed"].median(),inplace=True)
loan_train["Gender"].fillna(loan_train["Gender"].median(),inplace=True)
loan_train["Property_Area"].fillna(loan_train["Property_Area"].median(),inplace=True)

loan_test["Loan_Status"].fillna(loan_test["Loan_Status"].median(),inplace=True)
loan_test["Married"].fillna(loan_test["Married"].median(),inplace=True)
loan_test["Education"].fillna(loan_test["Education"].median(),inplace=True)
loan_test["Self_Employed"].fillna(loan_test["Self_Employed"].median(),inplace=True)
loan_test["Gender"].fillna(loan_test["Gender"].median(),inplace=True)
loan_test["Property_Area"].fillna(loan_test["Property_Area"].median(),inplace=True)
```

```python
loan_train['Loan_Status'].unique()
```

```python
lb_make = LabelEncoder()
loan_train["Loan_Status"] = lb_make.fit_transform(loan_train["Loan_Status"])
loan_train["Married"] = lb_make.fit_transform(loan_train["Married"])
loan_train["Education"] = lb_make.fit_transform(loan_train["Education"])
loan_train["Self_Employed"] = lb_make.fit_transform(loan_train["Self_Employed"])
loan_train["Gender"] = lb_make.fit_transform(loan_train["Gender"])
loan_train["Property_Area"] = lb_make.fit_transform(loan_train["Property_Area"])
loan_test["Loan_Status"] = lb_make.fit_transform(loan_test["Loan_Status"])
loan_test["Married"] = lb_make.fit_transform(loan_test["Married"])
loan_test["Education"] = lb_make.fit_transform(loan_test["Education"])
loan_test["Self_Employed"] = lb_make.fit_transform(loan_test["Self_Employed"])
loan_test["Gender"] = lb_make.fit_transform(loan_test["Gender"])
loan_test["Property_Area"] = lb_make.fit_transform(loan_test["Property_Area"])
```

```python
loan_train.head()
```

```python
loan_test.head()
```

```python
features = ['Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Credit_History','Property_Area']

x_train = loan_train.loc[:,features]
x_test = loan_test.loc[:,features]

y_train = loan_train.loc[:,['Loan_Status']]
y_test = loan_test.loc[:,['Loan_Status']]
```

```python
y_test.head()
```

```python
logreg = LogisticRegression(C=1e5,solver='lbfgs',multi_class='multinomial')
logreg.fit(x_train,y_train)
y_pred = logreg.predict(x_test)
```

```python
y_pred
```

```python
Y = []
for i in y_test.to_numpy():
    Y.append(i[0])
Y
```

```python
accuracy = np.sum(y_pred == Y)/len(y_pred)
print("%s: %f" % ("The accuracy on test data is", accuracy))
```

```python
for i in range(len(x_test)):
    print('LoanID : {} *** Probability to be Approved : {},   ***  Loan Aplication Result: {}'.format(loan_test['Loan_ID'][i], logreg.predict_proba(x_test)[i][1], y_pred[i]))
```
